```python
import pandas as pd # 引用套件並縮寫為 pd  
import numpy as np
```

```python
from keras import layers
from keras import models
from keras import regularizers
from keras.layers import Embedding,LSTM,Dense,SimpleRNN

def create_model():
    model=models.Sequential()
    model.add(Embedding(num_top_word,64))
    model.add(layers.Bidirectional(LSTM(64)))
    #model.add(LSTM(32, return_sequences=True))
    #model.add(LSTM(32, return_sequences=True))
    #model.add(LSTM(32, return_sequences=True))
    #model.add(LSTM(32))
    model.add(Dense(1,activation='sigmoid'))
    model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
    return model
```

```python
num_top_word=1000
maxlen=30
```

```python
data_train = pd.read_csv('data/train.csv')  
print(data_train) 
```

```python
data_test = pd.read_csv('data/test.csv')  
print(data_test) 
```

```python
data_train=data_train[['text','target']]
data_train['text']=data_train['text'].str.lower()
data_train
```

```python
from keras.preprocessing.text import Tokenizer
```

```python
data_train_=data_train['text']
```

```python
tokenizer=Tokenizer(num_words=num_top_word)
tokenizer.fit_on_texts(data_train_)
seq=tokenizer.texts_to_sequences(data_train_)
seq
```

```python
seq = np.array(seq)
```

```python
from keras.preprocessing import sequence
seq=sequence.pad_sequences(seq, maxlen=maxlen)
```

```python
seq.shape
```

## 建立model

```python
model=create_model()
model.summary()
```

```python
train_x=seq
train_y=data_train['target']
```

```python
train_x.shape
```

```python
history=model.fit(train_x,train_y,epochs=15,batch_size=128,validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
#plt.show()
plt.savefig("model_V4-1.png")
```

```python
plt.clf()
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
#plt.show()
plt.savefig("model_V4-2.png")
```
